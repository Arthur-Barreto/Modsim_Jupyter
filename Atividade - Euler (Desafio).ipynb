**item 1** 

Considere o seguinte sistema de equações diferenciais:

$\frac{dL}{dt}= k_1\cdot L\cdot(\frac{A}{20}-1)$

$\frac{dA}{dt}= k_2\cdot A\cdot(1- \frac{L}{10})$

A unidade de tempo utilizada para todos os parâmetros e fluxos é a **hora**.

Implemente na célula abaixo a solução numérica desse sistema usando o **o método de Euler**. Plote em um único gráfico as soluções $L(t)$ e $A(t)$. Para isso, utilize uma lista de tempo de 0 a 50h com $\Delta t=10^{-4}h$.

Os valores iniciais $L(0)$ e $A(0)$ devem ser:

- $L(0)=$ 20
- $A(0)=$ 35

Utilize os seguintes parâmetros: $k_1=0.2$ e $k_2=0.4$

```python
# seu código aqui para o item 1

import numpy as np
import matplotlib.pyplot as plt

delta_t = 1e-4
Tmax = 50
tempo = np.arange(0,Tmax, delta_t)

l = [0]*len(tempo)
a = [0]*len(tempo)

l[0] = 20
a[0] = 35

k1 = 0.2
k2 = 0.4

i = 1

while (i < len(tempo)):
    l[i] = l[i-1] + k1*l[i-1]*((a[i-1]/20) - 1)*delta_t
    a[i] = a[i-1] + k2*a[i-1]*(1 - (l[i-1]/10))*delta_t
    
    i = i + 1
    
plt.plot(tempo, l, 'b')
plt.plot(tempo, a, 'r')
plt.xlabel("Tempo em horas")
plt.ylabel("F(x)")
plt.title("Metodo de euler")
plt.grid(True)
plt.show()




    
```

**Item 2** 

Implemente agora o código que você criou no item 1 dentro de uma função denominada "PreencheSeriesTemporais". Essa função deve receber como parâmetro as constante $k_1$ e $k_2$ e retornar as listas A e L preenchidas. Usando a função, plote, num mesmo gráfico, a série de L para 4 valores distintos de $k_1$ e $k_2$.

```python
# seu código aqui para o item 2

def PreencheSeriesTemporais(k1,k2):

    delta_t = 1e-4
    Tmax = 50
    tempo = np.arange(0,Tmax, delta_t)

    l = [20]
    a = [35]

    i = 1

    while (i < len(tempo)):
        l.append(l[i-1] + k1*l[i-1]*((a[i-1]/20) - 1)*delta_t)
        a.append(a[i-1] + k2*a[i-1]*(1 - (l[i-1]/10))*delta_t)
        i = i + 1
        
    return l,a

l,a = PreencheSeriesTemporais(0.2,0.4)

plt.plot(tempo, l)

l,a = PreencheSeriesTemporais(0.4,0.8)

plt.plot(tempo, l)

l,a = PreencheSeriesTemporais(0.8,1.6)

plt.plot(tempo, l)

l,a = PreencheSeriesTemporais(1.6,3.2)

plt.plot(tempo, l)

plt.xlabel("Tempo em horas")
plt.ylabel("F(x)")
plt.title("Metodo de euler")
plt.grid(True)
plt.show()


    

```
