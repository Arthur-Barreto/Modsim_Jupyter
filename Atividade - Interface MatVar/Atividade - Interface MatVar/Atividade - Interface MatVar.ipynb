**Introdução** 

A cintilografia é um exame diagnóstico que avalia a função e a forma dos órgãos por meio da injeção de pequena quantidade de uma substância radioativa chamada radiofármaco. Essa substância emite radiações para as câmeras e computadores, que são transformadas em imagens coloridas, permitindo calcular a concentração e a velocidade de difusão da substância em diferentes partes do corpo.
Na cintilografia da tireoide, utiliza-se o I-131, um isótopo radioativo do iodo. Faz-se a aplicação intravenosa de certa quantidade, medida em $\mu Ci$ (micro Curie, ou seja, $10^{-6}$ Curie, sendo o Curie ($Ci$) uma unidade de medida de radioatividade), de iodeto de sódio radioativo ($NaI$). 
A partir daí, a concentração de $NaI$ é monitorada no sangue, na tireoide e no resto do corpo. Sabe-se que parte da substância inicialmente presente no sangue é absorvida pela tireoide, sendo o restante eliminado pela urina. O $NaI$ da tireoide passa, então, para o resto do corpo, de onde é eliminado pelas fezes ou devolvido ao sangue. Dessa forma, a modelagem do processo de cintilografia da tireoide pode ser feita considerando-se três compartimentos: sangue, tireoide e resto do corpo. 
Para simplificar a modelagem matemática, porém, vamos considerar apenas dois compartimentos, o sangue e a tireoide. Com isso, temos o seguinte diagrama de estoques e fluxos, em que todas as constantes indicadas são dadas em $dia^{-1}$.

Como você já deve saber, os códigos compilados no Jupyter Notebook rodam um na sequência do outro. Se você declara uma variável no começo, pode utilizar essa mesma variável mais adiante.

Dado isso, vamos já importar todas as bibliotecas que serão utilizadas nessa atividade, assim não precisamos ficar importando elas novamente. Para tal, basta rodar o código abaixo antes de ir para o próximo item.

```python
# Importa bibliotecas necessárias
from scipy.integrate import odeint
from math import *
import numpy as np
import matplotlib.pyplot as plt
```

**Item 1** 

Escreva em um papel as equações diferenciais do modelo. Após fazer isso, declare no espaço a seguir as constantes do problema e implemente a função que calcula as derivadas relevantes a partir dos parâmetros e variáveis do problema (ou seja, a função que você utilizará, posteriormente, na chamada de **ODEINT**).

**Atenção:** Não execute, ainda, a função **ODEINT**. Esse passo será feito no **Item 2**.

```python
# Escreva aqui a implementação do seu código do item 1
def derivada (x,t,kf,ks,kt,ku):
    S = x[0]
    T = x[1]
    dSdt = kt*T - ks*S - ku*S
    dTdt = ks*S - kt*T - kf*T
    dxdt = [dSdt,dTdt]
    return dxdt
    
```

**Item 2** 

Agora que você implementou a função que será usada por **ODEINT**, proceda aos seguintes passos na célula abaixo:

- Declare uma lista de tempo de $0$ a $10$ $dias$ com $\Delta t=10^{-3}$ $dia$

- Declare as condições iniciais do problema em uma lista. Use $S(0)=1$ e $T(0)=0$.

- Execute a função **ODEINT** e plote um gráfico das soluções numéricas $S(t)$ e $T(t)$.

```python
# Escreva aqui a implementação do seu código do item 2
########## 
##CI
S_0 = 1 
T_0 = 0
CI = [S_0,T_0]
##########
##EIXO X, TEMPO
inicio = 0
fim = 10
dt = 1e-3
t_lista = np.arange(inicio,fim,dt)
y_lista = odeint(derivada,CI,t_lista,args = (0.13,1,0.27,2))

#vamos separar as listas com as integrais
# coluna 0 é a primeira variável
# a coluna 1 é a segunda
S_lista = y_lista[:,0]
T_lista = y_lista[:,1]
#########
##PLOTAR OS GRÁFICOS
plt.plot(t_lista,S_lista,'r', label = "Radioatividade no Sangue")
plt.plot(t_lista,T_lista, 'k', label = "Radioatividade na Tireoide")
plt.ylabel("Radiatividade")
plt.xlabel("t")
plt.legend()
plt.title("Comportamento no Sangue")
plt.grid(True)
plt.show()
```

**Item 3** 

Agora, em vez de usar **ODEINT**, você vai resolver o sistema novamente, mas pelo método de Euler.

- Usando a mesma lista de tempo e as mesmas condições iniciais que você declarou no item anterior, implemente o método de Euler em um loop, usando **obrigatoriamente a função que você implementou no Item 1**. 

- Plote um gráfico das soluções numéricas $S(t)$ e $T(t)$ obtidas.

- Compare os dois resultados: Euler e Runge-Kutta (ou seja, o utilizado por **ODEINT**, que você obteve no **Item 2**).

```python
# Escreva aqui a implementação do seu código do item 3

##### lista de tempo pronta da celula anterior
# dt é dt

######### implementando as equações

kf = 0.13
ks = 1
kt = 0.27
ku = 2

S_euler = [1]
T_euler = [0]

for i in range(1,len(t_lista)):
    x = [S_euler[i-1],T_euler[i-1]]
    lista1 = derivada(x,t_lista[i-1],kf,ks,kt,ku)
    S_euler.append(S_euler[i-1] + lista1[0]*dt)
    T_euler.append(T_euler[i-1] + lista1[1]*dt)
    
plt.plot(t_lista,S_lista,'gx', label = "Radioatividade no Sangue")
plt.plot(t_lista,T_lista, 'bx', label = "Radioatividade na Tireoide")
plt.plot(t_lista,S_euler,'r', label = "Radioatividade no Sangue")
plt.plot(t_lista,T_euler, 'k', label = "Radioatividade na Tireoide")
plt.ylabel("Radiatividade")
plt.xlabel("t")
plt.legend()
plt.title("Comportamento no Sangue usando o metodo de euler")
plt.grid(True)
plt.show()

    

```

**Item 4** 

De modo geral, encontrar as soluções analíticas de um sistema de EDOs é uma tarefa bastante trabalhosa e, por vezes, impossível. Porém, existem situações em que isso é possível, como você verá com mais detalhes em Matemática da Variação. Uma dessas situações ocorre quando as expressões que descrevem as derivadas (no nosso caso, $\frac{dS}{dt}$ e $\frac{dT}{dt}$) são do primeiro grau nas variáveis dependentes (no nosso caso, $S$ e $T$).

Em casos como esse, a solução pode ser obtida por meio de conhecimentos de Álgebra Matricial (finalmente, as tais matrizes terão alguma utilidade!). Considerando os dados do nosso problema, as soluções analíticas são:

<center>$S(t)=\frac{27}{28}\cdot e^{-3,1\cdot t}+\frac{1}{28}\cdot e^{-0,3\cdot t}$</center>

<center>$T(t)=-\frac{10}{28}\cdot e^{-3,1\cdot t}+\frac{10}{28}\cdot e^{-0,3\cdot t}$</center>

Agora, você vai comparar essas soluções analíticas com a solução numérica obtida por meio de ODEINT no **Item 2**.
Para a implementação, crie duas listas (uma para $S(t)$ e outra para $T(t)$) com o mesmo tamanho da lista de tempo implementada no **Item 1** e, em seguida, preencha-as usando um loop. Note que, para plotar os gráficos da solução analítica, **não há equações a diferenças!**.

Plote os resultados e compare com os que você obteve por meio do método de Euler e usando **ODEINT**. São iguais? 

```python
# Escreva aqui a implementação do seu código do item 4

S_t = [1]
T_t = [0]

for i in range(0,len(t_lista)-1):
    S_t.append((27/28)*exp(-3.1*t_lista[i]) + (1/28)*exp(-0.3*t_lista[i]))
    T_t.append((-10/28)*exp(-3.1*t_lista[i]) + (10/28)*exp(-0.3*t_lista[i]))
    
plt.plot(t_lista,S_t,'r', label = "Radioatividade no Sangue")
plt.plot(t_lista,T_t, 'k', label = "Radioatividade na Tireoide")
plt.ylabel("Radiatividade")
plt.xlabel("t")
plt.legend()
plt.title("Solução Analitica")
plt.grid(True)
plt.show()
    

```

**Item 8 (para quem quer saber mais sobre a solução analítica)** 

Podemos reescrever as equações diferenciais do nosso sistema na forma matricial. Confira se a multiplicação das matrizes abaixo confere com as equações diferenciais que você descreveu no **Item 1**:

<center>
$\begin{bmatrix}
    \frac{dS}{dt}\\
    \frac{dT}{dt}    
\end{bmatrix}
=
\begin{bmatrix}
    -3 & 0,27\\
    1 & -0,4
\end{bmatrix}
\cdot
\begin{bmatrix}
    S\\
    T    
\end{bmatrix}$
</center>

Na equação matricial acima, temos uma matriz $2\times 2$ que relaciona as derivadas de $S$ e $T$ com as próprias variáveis $S$ e $T$. No código abaixo, estamos declarando essa matriz numa estrutura de dados do Python denominada **matrix**. A partir daí, executamos uma linha de código que resulta em dois valores. Note que esses valores são exatamente os fatores multiplicativos nos expoentes das soluções do **Item 4**. Esses são os famosos autovalores da matriz:

<center>
$\begin{bmatrix} -3 & 0,27\\1 & -0,4 \end{bmatrix}$
    </center>

Mas, o que são autovalores? Bom, isso você vai aprender em MatVar! Mantenha-se curioso(a) até lá!

```python
# Importa bibliotecas adicionais necessárias
from numpy import linalg
from numpy import matrix

#vamos iniciar a matriz A
A=matrix([[-3,0.27],[1,-0.4]])
print("Nossa matriz é:\n", A)

#vamos calcular e imprimir seus autovalores
autovalores=linalg.eig(A)[0]
print("Os autovalores da matriz são: %.2f" %(autovalores)[0], "e %.2f" %(autovalores)[1])
```
