Lembre-se que nossas equaes diferenciais podem ser escritas na forma (mais fcil de implementar):

$\frac{dT_o}{dt}=\frac{1}{m_o\cdot c_o}\cdot[\dot{Q_a}-\dot{Q_s}-\dot{Q_r}]$

$\frac{dT_r}{dt}=\frac{1}{m_r\cdot c_r}\cdot[\dot{Q_r}-\dot{Q_p}]$

Sendo:

$\dot{Q_a}=P$

$\dot{Q_s}=\frac{T_o-T_a}{\frac{1}{h_s\cdot A_s}}$

$\dot{Q_r}=\frac{T_o-T_r}{\frac{d_r}{k_r\cdot A_r}}$

$\dot{Q_p}=\frac{T_r-T_a}{\frac{d_r}{k_r\cdot A_r}+\frac{1}{h_p\cdot A_p}}$


Abaixo, inicializamos todos os parmetros do problema. Observe todas as unidades e valores que usaremos no nosso modelo. Seria interessante se voc procurasse por alguns desses valores como, por exemplo, o calor especfico do leo de soja, ou a condutividade trmica do ao ( mandatria a pesquisa em ingls, caso contrrio, vocs encontraro muito pouca informao). Se estiver confortvel com todos os parmetros, siga para o item 2.1. Aproveite, tambm, para preencher com comentrios as unidades que no foram indicadas. 

```python
#densidade do leo de soja - 0,917 kg/L --> 917 kg/m3
do=917
#densidade do recipiente (lato)
drec=8500
#volume de leo - vamos deixar 10cm abaixo da altura do recipiente
vo=1*1*0.5
#volume do recipiente
vr=0.05*1*1*5
#massa de leo
mo=do*vo
#massa do recipiente
mr=drec*vr
#calor especfico do leo de soja
co=1814.9
#calor especfico do lato
cr=377
#temperatura ambiente, em kelvin
Ta=26+273.15
#espessura do recipiente: 5cm --> 0.05m: vamos dividi-lo ao meio na modelagem da conduo trmica
dr=0.05/2
#rea da superfcie de contato entre o leo e o ar
As=1*1
#rea de contato das paredes externas do recipiente
Ap=0.6*1*4
#rea de contato das paredes internas recipiente (vamos simplificar e igual-las aqui)
Ar=Ap
#condutividade trmica do ao --> 80W/mk
kr=160
#coeficiente de transferncia convectiva da superfcie do leo para o ar --> 10W/m2k
hs=10
#coeficiente de transferncia convectiva das paredes do recipiente para o ar --> 10W/m2k
hp=10
#potncia trmica dos aquecedor --> 18kW
P=18000
#constante de Boltzman
sigma=5.6703e-8
#emissividade do oleo
emissividade = 0.95
```

**(2.1) Implementando a funo que representa as equaes diferenciais**

Voc se lembra de que, para usar a funo **ODEINT**,  necessrio que o sistema esteja escrito na forma abaixo?

<div>
     
</div>


A funo **ODEINT** exige que implementemos uma funo que represente o lado direito das igualdades do sistema acima. Sendo mais precisos, devemos implementar uma funo que, a partir de uma **lista** $x=[y_1, y_2, ..., y_N]$ e do instante $t$, retorne uma **lista** $\frac{dx}{dt}=\left[ \frac{dy_1}{dt}, \frac{dy_2}{dt}, ..., \frac{dy_N}{dt} \right]$.

No caso que estamos analisando, temos $N=2$, com $y_1=T_o$ e $y_2=T_r$, certo? Assim, devemos implementar uma funo que receba uma lista com os valores de $T_o$ e $T_r$, e o instante $t$, retornando uma lista com valor das derivadas $\frac{dT_o}{dt}$ e $\frac{dT_r}{dt}$.

Implemente essa funo no espao abaixo. Perceba que todos os parmetros necessrios para constru-la esto declarados acima. Se tiver alguma dvida sobre os fluxos, resistncias trmicas etc., consulte os vdeos da aula 17.


```python
#IMPLEMENTE AQUI SUA FUNO QUE SER USADA NO ODEINT
def modelo(x, t):
    To=x[0]
    Tr=x[1]
    Qa=P
    Qs=(To-Ta)/(1/(hs*As))
    Qr=(To-Tr)/(dr/(kr*Ar))
    Qp=(Tr-Ta)/((dr/(kr*Ar))+(1/(hp*Ap)))
    dTodt=(Qa-Qs-Qr)/(mo*co)
    dTrdt=(Qr-Qp)/(mr*cr)
    dxdt=[dTodt, dTrdt]
    return dxdt
```

**(3.1) Implementao da segunda iterao**

Implemente novamente, usando o espao abaixo, a funo que  utilizada pelo **odeint** para calcular as derivadas $\frac{dT_o}{dt}$ e $\frac{dT_r}{dt}$, agora levando em considerao a perda de calor do leo por radiao. Utilize a lei de Stefan Boltzmann, dada por $\dot{Q_{rad}}=e\cdot \sigma \cdot A\cdot T_o^4$. Aqui,  importante observar que a rea a ser usada na equao da radiao  apenas a rea da superfcie, pois as paredes de metal tm a propriedade de refletir a radiao emitida pelo leo.


```python
#IMPLEMENTE AQUI A SEGUNDA ITERAO DO MODELO

def modelo2(x, t):
    To=x[0]
    Tr=x[1]
    Qa=P
    Qs=(To-Ta)/(1/(hs*As))
    Qr=(To-Tr)/(dr/(kr*Ar))
    Qp=(Tr-Ta)/((dr/(kr*Ar))+(1/(hp*Ap)))
    Qrad=(As)*emissividade*sigma*To**4
    dTodt=(Qa-Qs-Qr-Qrad)/(mo*co)
    dTrdt=(Qr-Qp)/(mr*cr)   
    dxdt=[dTodt, dTrdt]
    return dxdt


#executando o odeint
x=odeint(modelo2, ci, lisTempo)
#converso de unidades
lisToCelsius=[temp-273.15 for temp in x[:,0]]
lisTrCelsius=[temp-273.15 for temp in x[:,1]]
lisTempoH=[t/3600 for t in lisTempo]
#plotando o grfico como pedido
plt.plot(lisTempoH, lisToCelsius, label='To Calculado')
plt.plot(lisTempoH, lisTrCelsius, label='Tr Calculado')
plt.plot(TempoExp, TemperaturaExp, 'ro', label='To Medido')
plt.ylabel("Temperatura - graus Celsius")
plt.xlabel("Tempo (h)")
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

**c)** Plote um grfico do tempo necessrio para atingir 190C (ou seja, no eixo das ordenadas teremos o tempo, em **horas**, e no eixo das abscissas, a potncia do queimador, em $kW$).
```python
#implementao da funo que ser usada no ODEINT
def modelo3(x, t, Paq):
    To=x[0]
    Tr=x[1]
    Qa=Paq
    Qs=(To-Ta)/(1/(hs*As))
    Qr=(To-Tr)/(dr/(kr*Ar))
    Qp=(Tr-Ta)/((dr/(kr*Ar))+(1/(hp*Ap)))
    Qrad=(As)*emissividade*sigma*To**4
    dTodt=(Qa-Qs-Qr-Qrad)/(mo*co)
    dTrdt=(Qr-Qp)/(mr*cr)   
    dxdt=[dTodt, dTrdt]
    return dxdt


def TempoAte190graus(listaTemperatura, listaTempo):
    #vamos iniciar com valor mximo
    tempobuscado=1000  
    for i, temperatura in enumerate(listaTemperatura):
        if temperatura >= 190:
            tempobuscado=listaTempo[i]
            break
    return tempobuscado



lisP=[18e3, 25e3, 30e3, 40e3, 50e3]
lisTemposAte190=[]

for Pot in lisP:
    #executando o odeint
    Solucao=odeint(modelo3, ci, lisTempo, args=(Pot,))
    #converso de unidades
    lisTempCelsius=[temp-273.15 for temp in Solucao[:,0]]
    lisTempoH=[t/3600 for t in lisTempo]
    #plotando o grfico como pedido
    plt.plot(lisTempoH, lisTempCelsius, label=("P=%.1f kW" %(Pot/1000)))
    #encontra o tempo necessrio para atingir 190 graus
    lisTemposAte190.append(TempoAte190graus(lisTempCelsius, lisTempoH))

    
plt.ylabel("Temperatura do leo - graus Celsius")
plt.xlabel("Tempo (h)")
plt.legend(loc='best')
plt.grid(True)
plt.show()
  
    
#plotando o grfico conclusivo
ax=plt.axes()
ax.set_facecolor('xkcd:ivory')
plt.title("Grfico conclusivo")
plt.plot(lisP, lisTemposAte190, 'r-o')
plt.xlabel("Potncia do aquecedor (W)")
plt.ylabel("Tempo para atingir 190")
plt.grid(True)
plt.show()
```
