#### Item 1

Dada a seguinte equação diferencial de 2ª ordem: 

<center>$\dfrac{d^2y(t)}{dt^2}=2$</center>

Transforme esta equação diferencial de 2ª ordem em um sistema de equações diferenciais de 1ª ordem e implemente uma função `modelo` com elas.

Em seguida, utilize a função `odeint` para resolver este sistema de equações diferenciais com o $t$ variando de $0$ a $2,5$, para isso considere as seguintes condições inicias:
<center>$y(0)=0$</center>
<center>$\dfrac{dy(0)}{dt}=0$</center>

Por fim, plote os gráficos de $y(t)$ por $t$ e $\dfrac{dy(t)}{dt}$ por $t$ em dois gráficos distintos.

```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt

# Implemente seu código do Item 1 abaixo

def modelo(x,t):
    # a primeira posição de x vamos chamar da primeira derivada
    # a segunda posiçao de x vamos chamar da segunda derivada
    y = x[0] 
    v = x[1]
    dydt = v
    dvdt = 2
    dxdt = [dydt,dvdt]
    return dxdt

##########agora a parte com odeint#############

delta_t = 1e-4
inicio = 0
fim = 2.5

# dydt e y == 0, ci = 0 para ambos
ci = [0,0]
# lista de tempo para a odeint
lisTempo = np.arange(inicio,fim,delta_t)
# calculando v, a primeira integral
solucao=odeint(modelo, ci, lisTempo)

y = solucao[:,0]
v = solucao[:,1]

plt.plot(lisTempo, v, label="v(t)")
plt.xlabel("Tempo (s)")
plt.ylabel("Velocidade")
plt.legend()
plt.grid(True)
plt.show()

plt.plot(lisTempo, y, label="y(t)")
plt.xlabel("Tempo (s)")
plt.ylabel("Espaço")
plt.legend()
plt.grid(True)
plt.show()

```

#### Item 2

Dada a seguinte equação diferencial de 2ª ordem: 

<center>$\dfrac{d^2 y(t)}{dt^2}+\dfrac{dy(t)}{dt}+y(t)=0$</center>

Transforme esta equação diferencial de 2ª ordem em um sistema de equações diferenciais de 1ª ordem e implemente uma função `modelo` com elas.

Em seguida, utilize a função `odeint` para resolver este sistema de equações diferenciais com o $t$ variando de $0$ a $10$, para isso considere as seguintes condições inicias:
<center>$y(0)=1$</center>
<center>$\dfrac{dy(0)}{dt}=2$</center>

Por fim, plote os gráficos de $y(t)$ por $t$.

```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt

# Implemente seu código do Item 2 abaixo

def modelo(x,t):
    # a primeira posição de x vamos chamar da primeira derivada
    # a segunda posiçao de x vamos chamar da segunda derivada
    y = x[0] 
    z = x[1]
    dydt = z
    dzdt = -z - y
    dxdt = [dydt,dzdt]
    return dxdt

delta_t = 1e-4
inicio = 0
fim = 10

lisTempo = np.arange(inicio,fim,delta_t)
# a primeira posição representa y e a segunda y'
# do enunciado, y(0) = 0 e y'(0) = 2
ci = [1,2]
solucao=odeint(modelo, ci, lisTempo)
y = solucao[:,0]

plt.plot(lisTempo, y)
plt.xlabel("Tempo (s)")
plt.ylabel("y(t)")
plt.grid(True)
plt.show()
```
