**1) Implementação**

A partir do sistema do saltador definido em aula, plote em um mesmo gráfico a altura do anel ($y_a$) e da barra ($y_b$) em função do tempo com as seguintes condições iniciais:
- $y_a(0) = 0$
- $y_b(0) = h$
- $v_a(0) = 0$
- $v_b(0) = 0$

Considere os seguintes valores de aceleração da gravidade ($g$), massa do anel ($m_a$), massa da barra ($m_b$), altura da barra ($h$), raio do anel ($r_a$), raio da barra ($r_b$), comprimento natural (quando não esticado) do elástico ($l_0$) e constante elástica do elástico ($k$):

- $g = 9,81$ m/s$^2$
- $m_a = 0,01$ kg
- $m_b = 0,008$ kg
- $h = 0,1$ m
- $r_a = 0,05$ m
- $r_b = 0,005$ m
- $l_0 = 0,04$ m
- $k = 3$ N/m

**Observação:** execute a simulação por 3 segundos.

```python
# Importa bibliotecas
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# SEU CÓDIGO

g = 9.81
ma = 1e-2
mb = 8e-3
h = 1e-1
ra = 5e-2
rb = 5e-3
l0 = 4e-2
k = 3

lisTempo = np.arange(0,3,1e-2)

def modelo(x, t):
    # Desagrupa lista
    ya = x[0]
    yb = x[1]
    va = x[2]
    vb = x[3]    
    # Calcula as taxas de variação
    dyadt = va
    dybdt = vb
    l = math.sqrt((ra-rb)**2 + (yb - ya)**2)
    dvadt = ((1/ma)*(3*k*(l-l0))*((yb-ya)/l)) - g
    dvbdt = -((1/mb)*(3*k*(l-l0))*((yb-ya)/l)) - g
    # Verifica se a esfera já atingiu o chão (caso positivo para de integrar)
    if yb <= h and dvbdt <0:
        dybdt = 0
        dvbdt = 0  
    # Agrupe lista com as taxas de variação
    lista = [dyadt, dybdt, dvadt, dvbdt]
    # Retorna lista com as taxas de variação
    return lista

# condições iniciais
X0 = [0,h,0,0]
# executando odeint    
altura = odeint(modelo, X0, lisTempo)
# retornando ya e yb
ya = altura[:,0]
yb = altura[:,1]
va = altura[:,2]
vb = altura[:,3]

#Plotando o gráfico
plt.plot(lisTempo,ya, label = "Anel")
plt.plot(lisTempo,yb, label = "Barra")
plt.legend()
plt.title("Modelo saltador")
plt.ylabel("Altura (m)")
plt.xlabel("Tempo (s)")
plt.grid(True)
plt.show()
```

**2) Uma forma de validação**

Uma forma interessante de validar a sua implementação é verificar a variação da energia do sistema. A rigor, não estaremos validando o modelo em comparação com um caso real, mas verificaremos se ele é fisicamente coerente.

Como não foi considerado nenhum tipo de dissipação de energia, a energia mecânica ($E_m$) deve permanecer constante. Sabendo que a mesma é a somatória das energias cinética ($E_c$), potencial gravitacional ($E_{pg}$) e potencial elática ($E_{pe}$):
- $E_m = E_c+E_{pg}+E_{pe}$

E que cada uma delas é dada por:
- $E_c = \frac{1}{2} m_a \cdot v_a^2 + \frac{1}{2} m_b \cdot v_b^2 $
- $E_{pg} = m_a \cdot g \cdot y_a + m_b \cdot g \cdot y_b$
- $E_{pe} = 3 \left[ \frac{1}{2} k {\left( \sqrt{(r_b-r_a)^2+(y_b-y_a)^2}-l_0 \right)}^2 \right] $

Plote em um mesmo gráfico todas as energias em função do tempo e verifique se a energia mecânica de fato permanece constante.

```python
# SEU CÓDIGO
ec = []
epg = []
epe = []
em = []

for i in range(0,len(lisTempo)):   
    ec.append(0.5*ma*va[i]**2 + 0.5*mb*vb[i]**2)
    epg.append(ma*g*ya[i] + mb*g*yb[i])
    l = math.sqrt((ra-rb)**2 + (yb[i] - ya[i])**2)
    epe.append(3*(0.5*k*(l-l0)**2))
    em.append(ec[i] + epg[i] + epe[i])
    
plt.plot(lisTempo,em, label = "Em")
plt.plot(lisTempo,ec, label = "Ec")
plt.plot(lisTempo,epg, label = "Epg")
plt.plot(lisTempo,epe, label = "Epe")
plt.legend()
plt.title("Validação do modelo")
plt.ylabel("Energia mecânica")
plt.xlabel("Tempo (s)")
plt.grid(True)
plt.show()
```
